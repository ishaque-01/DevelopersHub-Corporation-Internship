# Exploring and Visualizing a simple dataset (Iris)

```python
# import useful libraries for reading datasets and visualizing them
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading dataset form seaborn package/library
df = sns.load_dataset("iris")
```

## Getting some summary of the dataset

```python
# printing 1st 5 instances of the dataframe
df.head()
```

```python
# total rows and columns in the dataframe
df.shape
```

```python
# Columns names in dataframe
df.columns
```

## Visualizing the dataset

```python
plt.scatter(df['sepal_length'], df['sepal_width'])
plt.title('Scatter plot of Sepal Length and Sepal Width')
plt.show()
```

```python
plt.scatter(df['petal_length'], df['petal_width'])
plt.title('Scatter plot of Petal Length and Petal Width')
plt.show()
```

```python
plt.hist(df['petal_length'])
plt.show()
```

```python
plt.hist(df['petal_width'])
plt.show()
```

```python
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Iris Data Distribution')
plt.show()
```
